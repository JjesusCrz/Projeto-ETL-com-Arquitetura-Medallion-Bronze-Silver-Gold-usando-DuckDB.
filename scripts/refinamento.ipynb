IMPORTANDO BIBILHOTECAS

```python
import duckdb
```

CRIANDO CONEX√ÉO COM O BANCO DE DDADOS - SILVER

```python
con = duckdb.connect(database='dados_duckdb.db', read_only= False)
```

CONSUMINDO TABELA PRODUTOS NA CAMADA SILVER

```python
df = con.execute("SELECT * FROM produtos").fetchdf()
df.head(10)
```

TRATANDO A TABELA

```python
df2 = df.drop(columns=['id_categoria','id_fornecedor'])
df2.head(10)
```

CRIANDO TABELA dim_produtos na camada GOLD e inserindo valores tratados

```python
con.execute("""
        CREATE TABLE IF NOT EXISTS dim_produtos(
        id BIGINT,
        nm_produto TEXT,
        vl_preco FLOAT 
        
        )

""")
```

```python
con.execute("INSERT INTO dim_produtos SELECT * FROM df2")
```

```python
resultado = con.execute("SELECT * FROM dim_produtos").fetchdf()
resultado.head(10)
```

```python
con.close()
```
